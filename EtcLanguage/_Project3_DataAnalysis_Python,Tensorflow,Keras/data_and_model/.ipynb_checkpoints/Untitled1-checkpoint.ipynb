```python
import pymongo
from pymongo import MongoClient
from mtranslate import translate

conn = MongoClient('192.168.0.5', 27017)
instar = conn.instar
datas = instar.tagdetail


total = []
for data in datas.find({"trans_tagapi":None}, {'_id':1,'tags':1, 'img':1, 'tag':1, 'view':1, 'love':1, 'tags_cnt':1}):
    total.append({"id":data['_id'], 
                  "tags":data["tags"], 
                  "img":data['img'],
                  "tag":data['tag'],
                  "view":data['view'], 
                  "love":data['love'],
                  "tags_cnt":data['tags_cnt']})
```

```python
len(total)
```

```python
def api_trans_db(input_tag):
    for data in input_tag:
        for url in data['img']:
            try:
                urllib.request.urlretrieve(url, 'google.png')
            except:
                continue
            img_base64 = img_to_base64('google.png')

            result = request_cloud_vison_api(img_base64)

            text_result=[]
            try:
                count = len(result["responses"][0]["labelAnnotations"])
            except:
                continue
            for i in range(0,count):
                text_r = result["responses"][0]["labelAnnotations"][i]["description"]
                text_result.append(text_r)

            pre_trans = ",".join(text_result)
            trans_data = translate(pre_trans, 'ko', 'en').split(",")   

            datas.update({"_id":data['id']}, {"img":data["img"],
                                              "tags":data["tags"],
                                              "tag":data['tag'],
                                              "trans_tagapi":trans_data,
                                              "view":data['view'],
                                              "love":data['love'],
                                              "tags_cnt":data['tags_cnt']})
        print("변환 완료")
```

```python
api_trans_db(total)
```

```python
from multiprocessing import Pool

pool = Pool(processes=4)
pool.map(api_trans_db, total)
```

### DB connect & Get data from DB

```python
trans_data = []
tag_data = []

for data in datas.find({}, {'_id':1, 'tagapi':1, 'tagbasic':1, "trans_tagapi":1}):
    trans_data.append([trans.strip(" ") for trans in data['trans_tagapi']])
    tag_data.append([tag[1:] for tag in data['tagbasic']])
```

```python
import json, codecs
data = {"img":trans_data, "tag":tag_data}
with open('trans_data.txt', 'wb') as f:
    json.dump(data, codecs.getwriter('utf-8')(f), ensure_ascii=False)
```

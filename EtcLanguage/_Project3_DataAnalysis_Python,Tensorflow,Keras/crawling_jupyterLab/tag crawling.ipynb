```python
df = pd.read_csv('kor.txt', encoding='utf-8') # ê°€-í£ì˜ ê°’.
```

```python
driver = webdriver.Chrome('chromedriver_win32/chromedriver.exe') # í¬ë¡¬ ì‹¤í–‰.
driver.implicitly_wait(3) # ì‹¤í–‰ ëŒ€ê¸° ì‹œì”.
driver.get('https://www.instagram.com/explore/tags/ìº í¼ìŠ¤/') # ë§¨ ì²˜ìŒ ë“œë¼ì´ë²„ ì‹¤í–‰ í›„ ì‹œì‘í•˜ëŠ” Page.
```

```python
tagname = [] # tagname ì €ì¥.
tagcount = [] # tagcount ì €ì¥.

f = open('hashtag_v2.txt', 'w', encoding='utf-8') # í¬ë¡¤ë§í•œ Tag ì €ì¥í•˜ê¸° ìœ„í•œ Open.

count = 0
for w in df:
    delay = 3 # seconds, delay ì‹œê°„.
    try:
        driver.find_element_by_xpath('//div[@class="LWmhU _0aCwM"]/input').send_keys('#'+ w ) # #(ìƒµ) + ê°€~ í£ ê¹Œì§€ íƒœê·¸ ê²€ìƒ‰.
        div = WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, '//div[@class="fuqBx"]'))) # div í´ë˜ìŠ¤ ì•ˆì— ê°’ì„ ê°€ì ¸ì™€ divì— ì €ì¥.
#         print(count)
        html = div.get_attribute('innerHTML') # divì— ìˆëŠ” HTML ê°’ ê°€ì ¸ì˜´.
        soup = BeautifulSoup(html, 'html.parser') # beautifulsoupìœ¼ë¡œ htmlê°’ ë°”ê¿”ì¤Œ.
        
        
        for a in soup.find_all('a'): # soupì—ì„œ a tagê°’ì„ ê°€ì ¸ì˜´.
            #a.find('span',{'class':'Ap253'})
            tagname.append(a.find('span',{'class':'Ap253'}).text) # tagNameë³€ìˆ˜ì— append.
            tagcount.append(a.select('div > div > div.Fy4o8 > span > span')[0].text) # tagcountë³€ìˆ˜ì— append.
        print(tagname)
        print(tagcount)

        WebDriverWait(driver, delay).until(EC.presence_of_element_located((By.XPATH, "//div[@class='aIYm8 coreSpriteSearchClear']")))  # instargramì— ìˆëŠ” ê²€ìƒ‰ X í´ë¦­.
        driver.find_element_by_xpath("//div[@class='aIYm8 coreSpriteSearchClear']").click()
        
        count += 1
    except TimeoutException:
        print("Loading took too much time!")
    finally:
        pass
```

```python
result = {'tagname':tagname, 'tagcount':tagcount} # ì €ì¥ëœ ê°’ë“¤ resultì— dictìœ¼ë¡œ ì €ì¥.
f.write(str(result)) # ë¬¸ìì—´ë¡œ resultê°’ íŒŒì¼ì— ì“°ê¸°.
f.close() # file ë‹«ê¸°.
```

```python
import numpy as np
import matplotlib.pyplot as plt
# jupyter
에서 쓸 때만 사용
%matplotlib inline
plt.rc('font', family='malgun gothic')
```

```python
data = np.arange(1, 11, 1) # 1부터 11까지 1씩 증가하는.
plt.plot(data, 'r') # 뒤에 r은 선 색깔 변경.
plt.show()
```

```python
x = np.arange(10)
y = np.sin(x) # x안에 들어가 있는 값에 대한 sin 값.
print(x, y)
#plt.plot(x, y, 'bo') # 앞에는 색깔 뒤에는 모양. bo는 블루색깔에 동그라미, b^는 블루 색깔에 삼각형.
#plt.plot(y, 'r+')
#plt.plot(x, y, 'go--') # --는 선이 그려짐.
plt.plot(x, y, 'go--', linewidth=5, markersize=20) # 라인 길이, 마커 사이즈 키움.
plt.show()
```

```python
x = np.arange(0, np.pi * 3, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x축')
plt.ylabel('x축')
plt.title('사인과 코사인')
plt.show(['sine','consine'])
```

```python
# subplot
plt.subplot(2,1,1)
plt.plot(x, y_sin, linewidth=2)
plt.title('sine')

plt.subplot(2,1,2)
plt.plot(x, y_cos, 'r', linewidth=2)
plt.title('consin')
plt.show()
```

```python
index = [0, 20, 40, 60, 80, 100]
kor = [0, 51, 33, 88, 44, 66]
eng = [60, 55, 53, 48, 24, 86]
math= [80, 51, 53, 38, 80, 96]
k, = plt.plot(index, kor, 'ro-')
e, = plt.plot(index, eng, 'b^-')
m, = plt.plot(index, math, 'gs-')

#이미지 저장
fig = plt.gcf()

plt.title("세 과목점수")
plt.legend([k,e,m],['국어', '영수', '수학'], loc= 4)
plt.grid(True)
plt.show()
fig.savefig('test.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# jupyter에서 쓸 때만 사용
%matplotlib inline
plt.rc('font', family='malgun gothic')
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
fig.add_subplot(2,2,3)

# random으로 난수 발생하여 차트 그리기
plt.plot(np.random.randn(50).cumsum(), 'ro--', label='good')
ax1.hist(np.random.randn(100), bins=20, color='b', alpha = 0.3)
ax2.scatter(np.arange(30), np.arange(30) + np.random.randn(30)) # 산포도.
```

```python
data = [60, 80, 100, 70]
plt.bar(range(len(data)), data)
plt.show()
plt.barh(range(len(data)), data)
plt.show()
plt.pie(data)
plt.show()
```

```python
from pandas import Series, DataFrame
sdata = Series(np.random.randn(10).cumsum(), index = np.arange(0, 100, 10)) # Series 가지고 데이터 만든 후 그래프 출력
print(sdata)
plt.plot(sdata)

fdata = DataFrame(np.random.randn(1000, 4), columns=list('abcd'))
#fdata = DataFrame(np.random.randn(1000, 4), index = pd.date_range('1/1/2000', periods=1000), columns=list('abcd'))
#print(fdata)
df = fdata.cumsum() # cumsum() : 누적합.
print(fdata.head())
plt.plot(df)
#plt.show()
```

```python
iris_data = pd.read_csv('testdata/iris.csv')
#print(iris_data.info())
print(iris_data.head(3))

plt.scatter(iris_data['Sepal.Length'], iris_data['Petal.Length'])
plt.show()
```

```python
iris_col = iris_data.loc[:, 'Sepal.Length':'Petal.Width'] # 칼럼 4개 전부 잡아옴.
print(iris_col)
pd.plotting.scatter_matrix(iris_col, diagonal='kde') # 밀도값 출력.
plt.show()
```

```python
cols = []
print(iris_data['Species'].unique())

for s in iris_data['Species']:
    choice = 0
    if s == 'setosa':
        choice = 1
    elif s == 'versicolor':
        choice = 2
    elif s == 'virginica':
        choice = 3
    cols.append(choice)
    
print(cols)
plt.scatter(iris_data['Sepal.Length'], iris_data['Petal.Length'], c = cols)
plt.xlabel('Sepal.Length')
plt.ylabel('Petal.Length')
plt.show()
```
